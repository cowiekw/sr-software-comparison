```python
import openpyxl
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
from matplotlib.colors import ListedColormap
```

```python
df = pd.read_csv('data/features_classes_sorted.csv')
del df['total_count']
df = df.set_index('name')
df.head()
```

## Reshape the dataframe

```python
long_df = pd.melt(df, var_name='feature', value_name='count', ignore_index=False)
long_df = long_df.reset_index()
long_df.head()
```

```python
### Plot a Stacked Bar Chart
```

```python
ax = sns.histplot(long_df, y='name', hue='feature', weights='count',
             multiple='stack', palette='tab20c')
plt.ylabel('Sytematic Review Software')
plt.xlabel('Feature Count')
plt.title('Feature Count')
```

```python
plt.savefig('output/stacked_bar.png')
```
